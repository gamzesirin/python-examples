```python
%pip install requests
%pip install beautifulsoup4
!pip install openpyxl
import requests # HTTP İsteği göndermemizi sağlar ve bize geri değer döndürür.

from bs4 import BeautifulSoup # Sayfa içerisindeki html dökümanını parse(ayrıştırmak) etmemizi sağlar
# ve ayrıştırdığımız sayfa içerisinde istediğimiz verileri arayıp kazımamızı sağlar.

import pandas as pd # Python'un olmazsa olmazı her projede katkısı olan pandasımız :)
```

```python
url = 'https://www.kariyer.net/is-ilanlari?kw=yazilim%20gelistirici' # Veri çekmek istediğimiz sayfanın url'sini değişkene atıyoruz.
```

```python
response = requests.get(url) # Url değişkenine atadığımız sayfaya GET methodu ile istek atıyoruz ve geri değer alıyoruz.
# print(response.status_code) # Aldığımız geri değeri kontrol edebiliriz eğer 200 çıktısını aldıysak sayfadan veri çekebileceğimiz anlamına geliyor.
```

```python
icerik = BeautifulSoup(response.text, 'html.parser') # BeautifulSoup kullanarak sayfanın HTML dökümanını parse ediyoruz
# parse ettiğimiz sayfanın içeriğini text ile alıyoruz ve icerik değişkenimize atıyoruz.
```

```python
basliklar = icerik.find_all("span", attrs={"class" : "k-ad-card-title multiline"}) # icerik değişkenine atadığımız sayfamızın içerisindeki ilan başlıklarını çekiyoruz
# bunun için çekmek istediğimiz elementin üzerine "element select tool" ile tıklayıp ulaşmamız gerekiyor.
# ulaştıktan sonra etiketin önce etiket adını örnek: <span></span> ise "span" yazıyoruz.
# ardından attributeslarına erişiyoruz yani <span class="title"></span> yani burda "class" : "title" şeklinde kullanmalıyız
# örnek : <div class="title">Iphone 11</div> Bu etiketten veri çekmek istersek ( "div", attrs={"class" : "title"} ) bu şekilde kullanırız.

firmalar = icerik.find_all("div", attrs={"class" : "subtitle"}) # **
lokasyon = icerik.find_all("div", attrs={"class" : "job-detail"}) # **
```

```python
ilanliste = [] # For döngüsünden dönecek olan verileri listeye eklemek için boş bir liste oluşturuyoruz 
firmaliste = [] # **
konumliste = [] # **
```

```python
for baslik in basliklar: # bs4 den gelen basliklar verisinin içinde for döngüsü ile dönüyoruz
    ilanlar = baslik.text # döngü ile baslik çıktısının yalnızca text değerlerini yani <div class="title">Yazılım Uzmanı</div> çıktısının sadece "Yazılım Uzmanı" kısmını alıyoruz.
    ilanliste.append(ilanlar) # ve yukarda oluşturduğumuz liste içerisine döngüden dönen her veriyi ekliyoruz.
```

```python
for firma in firmalar: # ** 
    firma = firma.text.strip()
    firmaliste.append(firma)
```

```python
for konum in lokasyon: # **
    konum = konum.text.strip().split("\n")[0]
    konumliste.append(konum)
```

```python
datas = {"Başlıklar" : ilanliste, "Firmalar" : firmaliste, "Lokasyonlar" : konumliste} # bir dictionary oluşturarak listelerimiz sözlüğe ekliyoruz bu sayede
# "Başlıklar" : Yazılım Uzmanı, "Firmalar" : Kodluyoruz Derneği, "Lokasyonlar": İstanbul
# çıktısı elde etmiş oluyoruz
```

```python
# ilanliste, firmaliste ve konumliste isimli listelerin uzunluklarını kontrol ediyoruz
max_len = max(len(ilanliste), len(firmaliste), len(konumliste))

# Listelerin uzunluklarını eşitleyerek eksik veya hatalı veri için yer tutucu değer ekliyoruz
ilanliste += ['N/A'] * (max_len - len(ilanliste))
firmaliste += ['N/A'] * (max_len - len(firmaliste))
konumliste += ['N/A'] * (max_len - len(konumliste))
```

```python
df = pd.DataFrame(datas) # oluşturduğumuz datas sözlüğünü pandas kütüphanesi ile dataframe formatına çeviriyoruz ve bunu df değişkenine atıyoruz.
```

```python
df = df[["Başlıklar" , "Firmalar", "Lokasyonlar"]] # oluşturduğumuz dataframe'e sütun adları veriyoruz
```

```python
df.to_excel("ilanlar.xlsx", sheet_name="İlanlar", index=False) # Son olarak to_excel fonksiyonu ile dataframe'i excel tablosuna export ediyoruz.
```
